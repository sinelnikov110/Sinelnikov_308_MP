```python
# 1, 3
# Plot the populations for the years in the period. Add legend and axis labels to the plot.
# Find which year each species had the largest population. Mark these years in the plot of populations.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# np.ma.where(x > 5, x, -3.1416)

x = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
years = x[:, 0]
hare = x[:, 1]
lynx = x[:, 2]
carrot = x[:, 3]


fig = plt.figure(figsize = (16,7), dpi=80)
fig.patch.set_facecolor('xkcd:white')


plt.plot(years, hare)
plt.scatter(years[np.argmax(hare)], np.max(hare))
plt.title('population hare', fontsize=16)
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.xticks(years ,fontsize=8)
plt.grid(False)


fig = plt.figure(figsize = (16,9), dpi=80)
fig.patch.set_facecolor('xkcd:white')

plt.plot(years, lynx)
plt.scatter(years[np.argmax(lynx)], np.max(lynx))
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.title('population lynx', fontsize=16)
plt.xticks(years ,fontsize=8)
plt.grid(False)



fig = plt.figure(figsize = (16,9), dpi=80)
fig.patch.set_facecolor('xkcd:white')

plt.plot(years, carrot)
plt.scatter(years[np.argmax(carrot)], np.max(carrot))
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.title('population carrot', fontsize=16)
plt.xticks(years ,fontsize=8)
plt.grid(False)



```

```python
# 2
# Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.
```

```python

fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')


plt.bar(years, hare)
plt.bar(years, np.where(hare > 50000, hare , 0))
plt.title('population hare', fontsize=16)
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.grid(False)


fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

plt.bar(years, lynx)
plt.bar(years, np.where(lynx > 50000, lynx , 0))
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.title('population lynx', fontsize=16)
plt.grid(False)


fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

plt.bar(years, carrot)
plt.bar(years, np.where(carrot > 50000, carrot , 0))
plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.title('population carrot', fontsize=16)
plt.grid(False)

```

```python
# 4
# Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.
```

```python
fig = plt.figure()
fig = plt.figure(figsize = (16,9), dpi=80)


plt.bar(years, hare, label = 'population hare')
plt.bar(years, lynx, label = 'population lynx')


plt.xlabel('year', fontsize=16)
plt.ylabel('population size', fontsize=16)
plt.legend()
plt.grid(False)
```

```python
# Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.
```

```python
fig, ax = plt.subplots(1, 3,  figsize=(25,8), sharey=True, sharex=True)

mean_hare=np.mean(hare)
mean_lynx=np.mean(lynx)
mean_carrot=np.mean(carrot)

std_hare=np.std(hare)
std_lynx=np.std(lynx)
std_carrot=np.std(carrot)


ax[0].bar(years, hare, label = 'population hare')
ax[0].hlines(mean_hare, years[0] , years[-1], colors = 'red', label = 'mean' )
ax[0].hlines(std_hare, years[0] , years[-1], colors = 'green', label = 'std' )



ax[1].bar(years, lynx, label = 'population lynx')
ax[1].hlines(mean_lynx, years[0] , years[-1], colors = 'red', label = 'mean')
ax[1].hlines(std_lynx, years[0] , years[-1], colors = 'green', label = 'std' )


ax[2].bar(years, carrot , label = 'population carrot')
ax[2].hlines(mean_carrot, years[0] , years[-1], colors = 'red', label = 'mean')
ax[2].hlines(std_carrot, years[0] , years[-1], colors = 'green', label = 'std' )


for i in range(3):
    ax[i].legend()
```

```python
# 6
# Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.
```

```python
meaning = list(map(lambda x: "H" if x else "L", hare > lynx))
meaning
```

```python
fig = plt.figure(figsize = (16,7), dpi=80)
fig.patch.set_facecolor('xkcd:white')

plt.pie(100*np.unique(meaning, return_counts=True)[1]/len(meaning), labels=['hare','lynx '], autopct='%1.1f%%',  shadow=True)
plt.show()
```

```python
# Plot the change in the hare population and the number of lynxes. Find the correlation coefficient.
```

```python
fig, ax = plt.subplots(1, 2,  figsize=(25,8), dpi=80, sharey=True, sharex=True)

Population_change_hare = np.zeros(hare.shape[0])
Population_change_hare[:-1] = hare[1:]-hare[:-1]

Population_change_lynx = np.zeros(lynx.shape[0])
Population_change_lynx[:-1] = lynx[1:]-lynx[:-1]


ax[0].plot(years, Population_change_hare)
ax[0].set_xticks(years, fontsize=5)

ax[1].plot(years, Population_change_lynx)
ax[1].set_xticks(years, fontsize=5)


plt.show()
```

```python
# Using a scatter plot, show the population of hares vs carrots and hares vs lynxes by year on two separate subplots (each point corresponds to a particular year, and the point coordinates correspond to the population of the two species in that year).
```

```python
fig, ax = plt.subplots(1, 2,  figsize=(10,5), sharey=True, sharex=True)

ax[0].scatter(hare, carrot)
ax[0].set_xlabel('hare', fontsize=20)
ax[0].set_ylabel('carrot', fontsize=20)

ax[1].scatter(hare, lynx)
ax[1].set_xlabel('hare', fontsize=20)
ax[1].set_ylabel('lynx', fontsize=20)


fig.tight_layout()
plt.show()
```
